```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(nrows=3, ncols=4, figsize=(20,12))
x = np.arange(1, 21, 1)
for i, case in enumerate(["best", "worst", "avr"]):
    for j, pow in enumerate(range(2, 6)):
        try:
            arr = np.load(f"./timings/{case}_pow{pow}.npy")
            ax[i, j].plot(x, arr)
            ax[i, j].set_title(f"{case} - N=10^{pow}")
        except:
            pass
```
